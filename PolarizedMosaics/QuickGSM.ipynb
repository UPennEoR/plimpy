```python
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
from pygsm import GlobalSkyModel2016, GlobalSkyModel
```

```python
maps = []
freqs = np.linspace(150,151,num=1)
for f in freqs:
    gsm = GlobalSkyModel(freq_unit='MHz')
    gsm.generate(f)
    maps.append(gsm.generated_map_data)
```

```python
#%matplotlib qt
hp.mollview(np.log10(maps[0]),title='', notext=True, coord = ['G','C'], max=3.5)
#plt.savefig('GSM150MHzEQ.png')
plt.show()
```

```python
hp.write_map('GSM150MHz.fits',maps[0])
```
