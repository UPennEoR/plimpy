```python
#bad antennas list
#puts good data points per antenna in gAntData, and lists bad antennas in bAnt 
#(currently, it is a bad antenna if more than less of data per antenna is good)

#also, makes a bad 10 min file list

gAntdata = np.zeros(len(uvc.ant_array))
bAnt=np.zeros(len(uvc.ant_array))

bFiledata=np.zeros(len(calfiles))
bFile=np.zeros(len(calfiles))

for idx,calfile in enumerate(calfiles):
    uvc.read_calfits(calfile)
    for ant in uvc.ant_array:
        if (uvc.flag_array[uvc.ant2ind(ant),0,:,:,1].sum()/(uvc.Nfreqs*uvc.Ntimes)) != 1.0:
            gAntdata[uvc.ant2ind(ant)] += 1
        else:    
            bFiledata[idx]+=1

for idx,gdata in enumerate(gAntdata):
    if gdata < (len(calfiles)-gdata):
        bAnt[idx]=1
        
for idx,bdata in enumerate(bFiledata):
    if bdata > (len(uvc.ant_array)-bdata):
        bFile[idx]=1
        
        
print(gAntdata)
########
#bad antennas (shown as 1)
print(bAnt)

########


print(bFiledata)

########
#bad 10 min files (shown as 1)
print(bFile)

######
```
