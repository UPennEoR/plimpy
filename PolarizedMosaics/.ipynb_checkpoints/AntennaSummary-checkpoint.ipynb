```python
#bad file list
#(currently, it is a bad antenna if the good data is under 50%)

bFiledata=np.zeros(len(calfiles))
bFile=np.zeros(len(calfiles))

antPercentage=np.zeros(len(uvc.ant_array))

for idx,calfile in enumerate(calfiles):
    uvc.read_calfits(calfile)
    for ant in uvc.ant_array:
        if (uvc.flag_array[uvc.ant2ind(ant),0,:,:,1].sum()/(uvc.Nfreqs*uvc.Ntimes)) == 1.0:   
            bFiledata[idx]+=1
            

for idx,bdata in enumerate(bFiledata):
    #currently at 50%
    if ((bdata / len(uvc.ant_array))*100) > 50:
        bFile[idx]=1


#list of number of bad data points per 10 min file
print(bFiledata)

# list of bad 10 min files, shown as ones, ordered numerically based on prefix
print(bFile)
```

```python
#list of bad files (given qualifications above)

bFilelist=[]
gFiles=[]
gprefixes=[]

for idx,prefix in enumerate(prefixes):
    print(prefix, bFile[idx])
    if bFile[idx] ==1.0:
        bFilelist.append(prefix)
        
#better list of bad 10 min files
print(bFilelist)


for idx,file in enumerate(bFile):
    if file == 0.:
        gFiles.append(calfiles[idx])
        gprefixes.append(prefixes[idx])
        
#number of good 10 min files
print(len(gFiles))
        
```

```python
%matplotlib notebook
```

```python
#plot of good data
#x-axis: each 10 min file (73)    y-axis: each antenna (52)

fig, ax1 = plt.subplots(figsize=(10,8))
ax1.title.set_text('Good Data Points')
#plt.figure(figsize=(8,7))
#plt.grid()

for prefix,gFile in zip(prefixes,gFiles):
    uvc.read_calfits(gFile)
    for ant in uvc.ant_array:
        if (uvc.flag_array[uvc.ant2ind(ant),0,:,:,1].sum()/(uvc.Nfreqs*uvc.Ntimes)) != 1.0:
            x=prefix
            y=uvc.ant2ind(ant)
            #y=ant
            ax1.plot(x, y, '.',color='black')
        #else:
        #    x=prefix
        #    y=uvc.ant2ind(ant)
        #    ax1.plot(x,y,'.',color='red')

ax1.set_xticklabels(gprefixes,rotation=90)
ax1.set_xlabel("10 min files",fontweight='bold',color='blue')
mn, mx = ax1.set_ylim(uvc.ant2ind(uvc.ant_array[0]), uvc.ant2ind(uvc.ant_array[-1]))
ax1.set_ylabel("Antennas (index numbers)",fontweight='bold',color='blue')


ax2 = plt.twinx()
ax2.set_ylim(uvc.ant_array[0],uvc.ant_array[-1])
#ax2.set_yticks(uvc.ant_array,minor=True)
ax2.set_ylabel("(antenna numbers)",fontweight='bold',color='blue')

#ax2.xticks(rotation=90)

```

```python
#plot of bad data pts

fig, ax1 = plt.subplots(figsize=(10,8))
ax1.title.set_text('Bad Data Points')

for prefix,gFile in zip(prefixes,gFiles):
    uvc.read_calfits(gFile)
    for ant in uvc.ant_array:
        if (uvc.flag_array[uvc.ant2ind(ant),0,:,:,1].sum()/(uvc.Nfreqs*uvc.Ntimes)) == 1.0:
            x=prefix
            y=uvc.ant2ind(ant)
            ax1.plot(x, y, '.',color='red')
            
ax1.set_xticklabels(gprefixes,rotation=90)
ax1.set_xlabel("10 min files",fontweight='bold',color='blue')
mn, mx = ax1.set_ylim(uvc.ant2ind(uvc.ant_array[0]), uvc.ant2ind(uvc.ant_array[-1]))
ax1.set_ylabel("Antennas (index numbers)",fontweight='bold',color='blue')


ax2 = plt.twinx()
ax2.set_ylim(uvc.ant_array[0],uvc.ant_array[-1])
ax2.set_ylabel("(antenna numbers)",fontweight='bold',color='blue')
```

```python
#bad antenna list


gAntdata = np.zeros(len(uvc.ant_array))
bAnt=np.zeros(len(uvc.ant_array))


for idx,gFile in enumerate(gFiles):
    uvc.read_calfits(gFile)
    for ant in uvc.ant_array:
        if (uvc.flag_array[uvc.ant2ind(ant),0,:,:,1].sum()/(uvc.Nfreqs*uvc.Ntimes)) != 1.0:
            gAntdata[uvc.ant2ind(ant)] += 1

for idx,gdata in enumerate(gAntdata):
    #currently 50%
    if ((gdata / len(gFiles))*100) < 50:
        bAnt[idx]=1
    antPercentage[idx]= (gdata / len(gFiles))*100
    
antPercentage[antPercentage == inf] = 0

#list of good antenna data per good 10 min file
print(gAntdata)

#list of bad antennas, shown as 1, in order of antenna indices
print(bAnt)
```

```python
#list of bad antennas (given qualifications above)

bAntlist=[]

for ant in uvc.ant_array:
    print(ant, bAnt[uvc.ant2ind(ant)])
    if bAnt[uvc.ant2ind(ant)]==1.0:
        bAntlist.append(ant)

#better list of bad antennas
print(bAntlist)
```

```python
#shows percentage, or reliability, of good data per antenna; based on each good 10 min file

for idx,ant in enumerate(uvc.ant_array):
    print(ant,antPercentage[idx],'%')
```
