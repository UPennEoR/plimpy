```python
%%time

import numpy as np
import matplotlib.pyplot as plt
from pyuvdata import UVData, UVCal, UVFlag, utils
import warnings
import glob
import os

warnings.filterwarnings(action='ignore')

JD = '2457548'

oldpath = '/lustre/aoc/projects/hera/plaplant/HERA19Golden/RawData/' + JD + '/'
newpath = '/lustre/aoc/projects/hera/jaguirre/HERA19Summer2020/RawData/' + JD + '/'
calpath = '/lustre/aoc/projects/hera/jaguirre/HERA19Summer2020/CalibratedData/' + JD + '/'

rawfiles = sorted(glob.glob(os.path.join(newpath, 'zen.' + JD +'*.uvcRP.uvh5')))
calfitsfiles = sorted(glob.glob(os.path.join(oldpath, 'zen.' + JD +'*.HH.uvcRP.calfits')))

uvd = UVData()
uvc = UVCal()
```

```python
print(uvc.gain_convention)
# V_calibrated = V_measured / g_i g*_j
print(uvc.gain_array.shape)
print(uvc.jpol2ind)
```

```python
%%time

for j in range(len(rawfiles[0:1])):
    
    uvd.read(rawfiles[j])
    uvc.read_calfits(calfitsfiles[j])
    
    print('\n', '\n', '\n', '\n', '\n', rawfiles[j])
    
    ants_actual = uvd.get_ants()
    
    
    #====================== plotting ======================#
    
    
    # plotting gains
    plt.figure(figsize=(18,10))
    plt.subplot(2,2,1)
    for i in np.arange(uvd.Nants_data):
        if uvc.flag_array[i,0,:,:,0].mean() != 1:
            plt.plot(uvc.freq_array.squeeze()/1e6,1./np.abs(uvc.gain_array[i,0,:,0,0]/~uvc.flag_array[i,0,:,0,0]),
                     label=ants_actual[i])
        #plt.legend()
    plt.xlim([100,200])
    plt.xlabel('Frequency (MHz)')
    plt.ylabel(r'(Jy/corr unit)$^{1/2}$')
    
    plt.subplot(2,2,2)
    for i in np.arange(uvd.Nants_data):
        if uvc.flag_array[i,0,:,:,0].mean() != 1:
            plt.plot(uvc.freq_array.squeeze()/1e6,1./np.abs(uvc.gain_array[i,0,:,0,1]/~uvc.flag_array[i,0,:,0,1]), 
                    label=ants_actual[i])
        #plt.legend()
    plt.xlim([100,200])
    plt.subplot(2,2,3)
    for i in np.arange(uvd.Nants_data):
        if uvc.flag_array[i,0,:,:,0].mean() != 1:
            plt.plot(uvc.freq_array.squeeze()/1e6,np.angle(uvc.gain_array[i,0,:,0,0]/~uvc.flag_array[i,0,:,0,0]),
                    label=ants_actual[i])
        plt.legend()
    plt.xlim([100,200])
    plt.subplot(2,2,4)
    for i in np.arange(uvd.Nants_data):
        if uvc.flag_array[i,0,:,:,0].mean() != 1:
            plt.plot(uvc.freq_array.squeeze()/1e6,np.angle(uvc.gain_array[i,0,:,0,1]/~uvc.flag_array[i,0,:,0,1]), 
                    label=ants_actual[i])
        plt.legend()
    plt.xlim([100,200])

    #plt.show()
    
    plt.savefig('gains.png')
    plt.close()
    #plt.subplot(1,2,2)
    #plt.plot(uvc.freq_array.squeeze()/1e6,np.angle(uvc.gain_array[1,0,:,0,0]/~uvc.flag_array[1,0,:,0,0]))
    #plt.plot(uvc.freq_array.squeeze()/1e6,np.angle(uvc.gain_array[1,0,:,0,1]/~uvc.flag_array[1,0,:,0,1]))
    #plt.xlim([115,185])
    
    
    #================ checking antennas =================#
    
    uvf_ant = UVFlag(uvc, mode='flag', copy_flags=True)
    for i, ant in enumerate(uvf_ant.ant_array):
        print('{:03} {:.5f} {:.5f}'.format(ant, uvf_ant.flag_array[i,0,:,:,0].mean(), 
                                        uvf_ant.flag_array[i,0,:,:,1].mean()))
        
    test_bls = [(88,64,'xx'), (97,10,'xx'), (20,31,'xx')]
    test_uvd = uvd.select(bls=test_bls, inplace=False)
    
    uvd_calibrated = utils.uvcalibrate(uvd, uvc, inplace=False)
    test_uvd_calibrated = uvd_calibrated.select(bls=test_bls, inplace=False)
    
    
    #================ plotting =================#
    
    
    # plotting visibilities
    plt.figure(figsize=(14,5))
    for i,app in enumerate(test_uvd.get_antpairpols()):
        plt.subplot(1,3,i+1)
        plt.title(app)
        plt.imshow(np.abs(test_uvd.get_data(app))/~test_uvd.get_flags(app),aspect='auto', 
                  vmin=0, vmax=1)
        plt.colorbar()
    plt.show()
    plt.figure(figsize=(14,5))
    for i,app in enumerate(test_uvd_calibrated.get_antpairpols()):
        plt.subplot(1,3,i+1)
        plt.title(app)
        plt.imshow(np.abs(test_uvd_calibrated.get_data(app))/~test_uvd_calibrated.get_flags(app),aspect='auto',
                  vmin=0, vmax=1200)
        plt.colorbar()
    plt.show()
    
    
    #================ plotting =================#
    
    plt.figure(figsize=(14,8))
    plt.subplot(2,1,1)
    for i,app in enumerate(test_uvd.get_antpairpols()):
        plt.plot((np.abs(test_uvd.get_data(app))/~test_uvd.get_flags(app))[0,:])
    plt.subplot(2,1,2)
    for i,app in enumerate(test_uvd_calibrated.get_antpairpols()):
        plt.plot((np.abs(test_uvd_calibrated.get_data(app))/~test_uvd_calibrated.get_flags(app))[0,:])
    plt.show()
    
    
    #================ plotting =================#
    
    fix_conj = [False, True, False]
    plt.figure(figsize=(14,5))
    for i,app in enumerate(test_uvd.get_antpairpols()):
        plt.subplot(1,3,i+1)
        plt.title(app)
        plt.imshow(np.angle(test_uvd.get_data(app))/~test_uvd.get_flags(app),aspect='auto')
        #plt.colorbar()
    plt.show()
    plt.figure(figsize=(14,5))
    for i,app in enumerate(test_uvd_calibrated.get_antpairpols()):
        plt.subplot(1,3,i+1)
        plt.title(app)
        if fix_conj[i]:
            vis = np.conj(test_uvd_calibrated.get_data(app))
        else:
            vis = test_uvd_calibrated.get_data(app)
        plt.imshow(np.angle(vis)/~test_uvd_calibrated.get_flags(app),aspect='auto')
    plt.show()

    #================ plotting =================#
    
    fix_conj = [False, True, False]
    plt.figure(figsize=(14,8))
    plt.subplot(2,1,1)
    for i,app in enumerate(test_uvd.get_antpairpols()):
        plt.plot((np.angle(test_uvd.get_data(app))/~test_uvd.get_flags(app))[0,:])
    plt.subplot(2,1,2)
    for i,app in enumerate(test_uvd_calibrated.get_antpairpols()):
        if fix_conj[i]:
            vis = np.conj(test_uvd_calibrated.get_data(app))
        else:
            vis = test_uvd_calibrated.get_data(app)
        plt.plot((np.angle(vis)/~test_uvd_calibrated.get_flags(app))[0,:])
    plt.show()
    
    #================ printing calibrated antennas =================#
    
    for i, bl in enumerate(uvd_calibrated.get_antpairpols()):
        print('{} {:.5f} {:.5f}'.format(bl,
                                    uvd_calibrated.get_flags(bl).mean(),
                                    uvd_calibrated.get_flags(bl).mean()))
    
    # getting the filename by itself without ending
    filename_only = calfitsfiles[j].split('/')[-1]
    split_filename = filename_only.split('.')
    
    
    #uvd_calibrated.write_uvfits(calpath+ stem[0] + '.' + stem[1] + '.' + stem[2] +'.uvcRP.calibrated.uvfits', 
                                #spoof_nonessential=True, force_phase=True)
    
    # writing calibrated uvfits file
    #uvd_calibrated.write_uvfits(split_filename[0] + '.' + split_filename[1] + '.' + split_filename[2] + '.uvcRP.calibrated.uvfits', 
    #                            spoof_nonessential=True, force_phase=True)
```
