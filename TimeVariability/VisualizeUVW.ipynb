```python
import numpy as np
import matplotlib.pyplot as plt
from pyuvdata import UVData
```

```python
path1 = '/lustre/aoc/projects/hera/jaguirre/PolarizedMosaics/2458098_sliced/'
path2 = '/lustre/aoc/projects/hera/jaguirre/PolarizedMosaics/2458099_sliced/'
```

```python
uvd1 = UVData()
uvd1.read(path1+'zen.092924_093331.calibrated.HH.uvfits')
```

```python
uvd2 = UVData()
uvd2.read(path2+'zen.092924_093331.calibrated.HH.uvfits')
```

```python
uvd1_uvws = np.reshape(uvd1.uvw_array,[uvd1.Ntimes, uvd1.Nbls, 3])
uvd2_uvws = np.reshape(uvd2.uvw_array,[uvd1.Ntimes, uvd1.Nbls, 3])
```

```python
plt.plot(uvd1_uvws[:,810,0])
plt.plot(uvd2_uvws[:,810,0])
```

```python
plt.plot(uvd1_uvws[:,810,2]-uvd2_uvws[:,810,2])
plt.plot(uvd1_uvws[:,650,2]-uvd2_uvws[:,650,2])
```

```python
for i in np.arange(3):
    plt.imshow(uvd1_uvws[:,:,i] - uvd2_uvws[:,:,i], aspect='auto')
    plt.colorbar()
    plt.show()
```
