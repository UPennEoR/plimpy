```python
def bl_length(uvd):
    """ Heavy forbid you want the baseline length """
    return np.sqrt(np.power(uvd.uvw_array, 2).sum(axis=1))
```

```python
def true_uvw(uv):
    """ Expand the uvw_array in meters to a proper one in wavelengths for all frequencies """
    
    uvw_m = np.repeat(np.expand_dims(uv.uvw_array, axis = 2), uv.Nfreqs, axis = 2)
    lmbda = ((c.c/(np.repeat(np.repeat(np.expand_dims(uv.freq_array, axis = 0), 3, axis = 1), 
                             uv.Nblts, axis = 0)*u.Hz) ).to(u.m)).value
    uvw = uvw_m / lmbda
    
    return uvw
```

```python
flags = ~uv8098.flag_array[:,0,:,0]
U = np.concatenate([uvw8098[:,0,:][flags].flatten(), -uvw8098[:,0,:][flags].flatten()])
V = np.concatenate([uvw8098[:,1,:][flags].flatten(), -uvw8098[:,1,:][flags].flatten()])
vis = np.concatenate([uv8098.data_array[:,0,:,0][flags].flatten(), 
                     np.conj(uv8098.data_array[:,0,:,0][flags]).flatten()])
```

```python
# uv gridding should be ... 1/FOV?  which depends on wavelength ...
# let's pick 
du = 1./((60*u.deg).to(u.rad).value)/4.
u_max = 80.
N_u = int(2.*np.ceil(u_max/du))

u_bins = np.linspace(-u_max, u_max, N_u)
v_bins = np.linspace(-u_max, u_max, N_u)
timer.start()
uv_psf, _ , _ = np.histogram2d(U, V, bins = [u_bins, v_bins])
timer.stop('Gridding uv coverage')
```

```python
timer.start()
uvvis_real, x_edge_, y_edge, binnumber = binned_statistic_2d(U, V, vis.real, statistic='mean', 
                                                           bins = [u_bins, v_bins])
uvvis_imag, x_edge_, y_edge, binnumber = binned_statistic_2d(U, V, vis.imag, statistic='mean', 
                                                           bins = [u_bins, v_bins])
timer.stop('Binning visibilities')
uvvis = uvvis_real + 1j*uvvis_imag
uvvis[~np.isfinite(uvvis)] = 0.
```

```python
uvvis[N_u/2-1,N_u/2-1] = 0.
```
